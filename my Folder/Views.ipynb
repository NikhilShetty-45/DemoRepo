```sql
use catalog hive_metastore;

create table if not exists smartphones
(id INT, name string, brand string, year INT);

Insert into smartphones
values (1, 'iPhone 12', 'Apple',2024),
       (2, 'Galaxy S23', 'Samsung',2024),
       (3, 'Pixel 7', 'Google',2024),
       (4, 'OnePlus 12', 'OnePlus',2024),
       (5 , 'Nokia', 'Microsoft',2024),
       (6, 'iPhone 13', 'Apple',2024),
       (7, 'Galaxy S24', 'Samsung',2024),
       (8, 'Pixel 8', 'Google',2024),
       (9, 'OnePlus 13', 'OnePlus',2024),
       (10, 'Nokia', 'Microsoft',2024);
```

```sql
SHOW TABLES
```

```sql
Create view view_apple_phones
as select * from smartphones where brand = 'Apple';
```

```sql
select * from view_apple_phones;
```

```sql
Show tables
```

```sql
create temp view temp_view_phone_brands
as select distinct brand
  from smartphones;
select * from temp_view_phone_brands
```

```sql
Show tables
```

```sql
create global temp view global_temp_view_phone_brands
as select * from smartphones
  where year > 2020
  order by year desc;
```

```sql
select * from global_temp.global_temp_view_phone_brands;
```
